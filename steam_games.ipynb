```python
import pandas as pd
import numpy as np
import os



```

```python
#Carga de archivo Json
steam = pd.read_json('../data/output_steam_games.json', lines= True)
steam.head()
```

```python
#Cuantos registros hay en la base
steam.shape[0]
```

```python
#Cuantos nulos hay en cada columna
steam.isnull().sum()
```

```python
#Eliminar columnas innecesarias
steam = steam.drop(['publisher','app_name','url','tags','reviews_url','early_access'], axis=1)
steam.head()
```

```python
#Convertir la columna price a numerico y rellenar vacios con la media de los datos
steam['price'] = pd.to_numeric(steam['price'], errors='coerce')
steam['price'].fillna(steam['price'].mean(), inplace=True)
```

```python
#Limpiar los registros que tengan nulos
steam_games_cleaned = steam.dropna()
steam_games_cleaned.head()
```

```python
#Cuantos datos quedaron despues del dropna
steam_games_cleaned.shape[0]
```

```python
# Convierte la columna 'id' a tipo de dato int
steam_games_cleaned['id'] = steam_games_cleaned['id'].astype(int)
# Renombra la columna 'id' como 'item_id'
steam_games_cleaned = steam_games_cleaned.rename(columns={'id': 'item_id'})
```

```python
# Desanidar los elementos de la columna genres
steam_games_cleaned = steam_games_cleaned.explode('genres')
```

```python
#Resetear el index
steam_games_cleaned = steam_games_cleaned.reset_index(drop=True)
steam_games_cleaned.head()
```

```python
# Utiliza la funcion explode en la columna 'specs' para convertir los valores anidados en filas separadas
steam_games_cleaned = steam_games_cleaned.explode('specs')
# Restablecer el índice del DataFrame resultante y descartar el índice anterior
steam_games_cleaned = steam_games_cleaned.reset_index(drop=True)
steam_games_cleaned.head()
```

```python
#Volver a revisar nulos
steam_games_cleaned.isnull().sum()
```

```python
# Elimina valores nulos de la columna genres
steam_games_cleaned = steam_games_cleaned.dropna(subset=['genres'])
# Elimina valores nulos de la columna id
steam_games_cleaned = steam_games_cleaned.dropna(subset=['item_id'])
# Elimina valores nulos de la columna specs
steam_games_cleaned = steam_games_cleaned.dropna(subset=['specs'])
```

```python
#Ponerle tipo string a los datos de la columna title
steam_games_cleaned['title'] = steam_games_cleaned['title'].astype(str)
```

```python
steam_games_cleaned.head()
```

```python
#Guardar en archivo cvs
steam_games_cleaned.to_csv(os.path.join('../data/', 'steam_games.csv'), index=False)
```
